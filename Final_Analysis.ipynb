```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
csvpath_chi_12 = os.path.join('2009,2019 Weather CSV', 'Chicago 2012.csv')
csvpath_chi_19 = os.path.join('2009,2019 Weather CSV', 'Chicago 2019.csv')
csvpath_LA_12 = os.path.join('2009,2019 Weather CSV', 'LAS 2012.csv')
csvpath_LA_19 = os.path.join('2009,2019 Weather CSV', 'LAS 2019.csv')
csvpath_ny_12 = os.path.join('2009,2019 Weather CSV', 'NY 2012.csv')
csvpath_ny_19 = os.path.join('2009,2019 Weather CSV', 'NY 2019.csv')
csvpath_philly_12 = os.path.join('2009,2019 Weather CSV', 'PHL 2012.csv')
csvpath_philly_19 = os.path.join('2009,2019 Weather CSV', 'PHL 2019.csv')
```

```python
# Read CSVs for each city and year into pandas and convert to Data Frame

with open(csvpath_chi_12) as csvfile_chi_12:
    csvreader_chi_12 = csv.reader(csvfile_chi_12, delimiter=",")
    csv_header = next(csvreader_chi_12) 
    chi_12_df_first = pd.DataFrame(csvreader_chi_12)
    
with open(csvpath_chi_19) as csvfile_chi_19:
    csvreader_chi_19 = csv.reader(csvfile_chi_19, delimiter=",")
    csv_header = next(csvreader_chi_19) 
    chi_19_df_first = pd.DataFrame(csvreader_chi_19)

with open(csvpath_LA_12) as csvfile_LA_12:
    csvreader_LA_12 = csv.reader(csvfile_LA_12, delimiter=",")
    csv_header = next(csvreader_LA_12) 
    LA_12_df_first = pd.DataFrame(csvreader_LA_12)

with open(csvpath_LA_19) as csvfile_LA_19:
    csvreader_LA_19 = csv.reader(csvfile_LA_19, delimiter=",")
    csv_header = next(csvreader_LA_19) 
    LA_19_df_first = pd.DataFrame(csvreader_LA_19)

with open(csvpath_ny_12) as csvfile_ny_12:
    csvreader_ny_12 = csv.reader(csvfile_ny_12, delimiter=",")
    csv_header = next(csvreader_ny_12) 
    ny_12_df_first = pd.DataFrame(csvreader_ny_12)

with open(csvpath_ny_19) as csvfile_ny_19:
    csvreader_ny_19 = csv.reader(csvfile_ny_19, delimiter=",")
    csv_header = next(csvreader_ny_19) 
    ny_19_df_first = pd.DataFrame(csvreader_ny_19)
    
with open(csvpath_philly_12) as csvfile_philly_12:
    csvreader_philly_12 = csv.reader(csvfile_philly_12, delimiter=",")
    csv_header = next(csvreader_philly_12) 
    philly_12_df_first = pd.DataFrame(csvreader_philly_12)

with open(csvpath_philly_19) as csvfile_philly_19:
    csvreader_philly_19 = csv.reader(csvfile_philly_19, delimiter=",")
    csv_header = next(csvreader_philly_19) 
    philly_19_df_first = pd.DataFrame(csvreader_philly_19)
```

```python
# Renaming original 2012 data frame columns and making a new column to store the Months calculated using a datetimeindex function on Date column 

chi_12_df = chi_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
chi_12_df['Month'] = pd.DatetimeIndex(chi_12_df['Date']).month

LA_12_df = LA_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
LA_12_df['Month'] = pd.DatetimeIndex(LA_12_df['Date']).month

ny_12_df = ny_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
ny_12_df['Month'] = pd.DatetimeIndex(ny_12_df['Date']).month

philly_12_df = philly_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
philly_12_df['Month'] = pd.DatetimeIndex(philly_12_df['Date']).month
chi_12_df.head()
```

```python
# Creating new 2012 Data Frames by grouping on the Month column, converting columns to numeric values, and then finding averages
chi_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = chi_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
chi_12_month = chi_12_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

LA_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = LA_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
LA_12_month = LA_12_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

ny_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = ny_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
ny_12_month = ny_12_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])
ny_12_month

philly_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = philly_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
philly_12_month = philly_12_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])
philly_12_month.head()
```

```python
# Renaming original 2019 data frame columns and making a new column to store the Months calculated using a datetimeindex function on Date column

chi_19_df = chi_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow",6:"Wind Speed"})
chi_19_df['Month'] = pd.DatetimeIndex(chi_19_df['Date']).month

LA_19_df = LA_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow",6:"Wind Speed"})
LA_19_df['Month'] = pd.DatetimeIndex(LA_19_df['Date']).month

ny_19_df = ny_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow",6:"Wind Speed"})
ny_19_df['Month'] = pd.DatetimeIndex(ny_19_df['Date']).month

philly_19_df = philly_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow",6:"Wind Speed"})
philly_19_df['Month'] = pd.DatetimeIndex(philly_19_df['Date']).month
philly_19_df.head()
```

```python
# Creating new 2019 Data Frames by grouping on the Month column, converting columns to numeric values, and then finding averages

chi_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = chi_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
chi_19_month = chi_19_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

LA_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = LA_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
LA_19_month = LA_19_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

ny_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = ny_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
ny_19_month = ny_19_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

philly_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = philly_19_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
philly_19_month = philly_19_df.groupby(['Month']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])
philly_19_month.head()
```
