```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
csvpath_chi_traffic_12 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'Chicago 2012 CrashReport.csv')
csvpath_chi_traffic_19 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'Chicago 2019 CrashReport.csv')
csvpath_LA_traffic_12 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'LAS 2012 CrashReport.csv')
csvpath_LA_traffic_19 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'LAS 2019 CrashReport.csv')
csvpath_ny_traffic_12 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'NY 2012 CrashReport.csv')
csvpath_ny_traffic_19 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'NY 2019 CrashReport.csv')
csvpath_philly_traffic_12 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'PHL 2012 CrashReport.csv')
csvpath_philly_traffic_19 = os.path.join('2012,2019 Vehicle Fatality CLEAN', 'PHL 2019 CrashReport.csv')
```

```python
with open(csvpath_chi_traffic_12) as csvfile_chi_traffic_12:
    csvreader_chi_traffic_12 = csv.reader(csvfile_chi_traffic_12, delimiter=",")
    csv_header = next(csvreader_chi_traffic_12) 
    chi_traffic_12_first = pd.DataFrame(csvreader_chi_traffic_12)
    
with open(csvpath_chi_traffic_19) as csvfile_chi_traffic_19:
    csvreader_chi_traffic_19 = csv.reader(csvfile_chi_traffic_19, delimiter=",")
    csv_header = next(csvreader_chi_traffic_19) 
    chi_traffic_19_first = pd.DataFrame(csvreader_chi_traffic_19)

with open(csvpath_LA_traffic_12) as csvfile_LA_traffic_12:
    csvreader_LA_traffic_12 = csv.reader(csvfile_LA_traffic_12, delimiter=",")
    csv_header = next(csvreader_LA_traffic_12) 
    LA_traffic_12_first = pd.DataFrame(csvreader_LA_traffic_12)

with open(csvpath_LA_traffic_19) as csvfile_LA_traffic_19:
    csvreader_LA_traffic_19 = csv.reader(csvfile_LA_traffic_19, delimiter=",")
    csv_header = next(csvreader_LA_traffic_19) 
    LA_traffic_19_first = pd.DataFrame(csvreader_LA_traffic_19)

with open(csvpath_ny_traffic_12) as csvfile_ny_traffic_12:
    csvreader_ny_traffic_12 = csv.reader(csvfile_ny_traffic_12, delimiter=",")
    csv_header = next(csvreader_ny_traffic_12) 
    ny_traffic_12_first = pd.DataFrame(csvreader_ny_traffic_12)

with open(csvpath_ny_traffic_19) as csvfile_ny_traffic_19:
    csvreader_ny_traffic_19 = csv.reader(csvfile_ny_traffic_19, delimiter=",")
    csv_header = next(csvreader_ny_traffic_19) 
    ny_traffic_19_first = pd.DataFrame(csvreader_ny_traffic_19)

with open(csvpath_philly_traffic_12) as csvfile_philly_traffic_12:
    csvreader_philly_traffic_12 = csv.reader(csvfile_philly_traffic_12, delimiter=",")
    csv_header = next(csvreader_philly_traffic_12) 
    philly_traffic_12_first = pd.DataFrame(csvreader_philly_traffic_12)

with open(csvpath_philly_traffic_19) as csvfile_philly_traffic_19:
    csvreader_philly_traffic_19 = csv.reader(csvfile_philly_traffic_19, delimiter=",")
    csv_header = next(csvreader_philly_traffic_19) 
    philly_traffic_19_first = pd.DataFrame(csvreader_philly_traffic_19)
```

```python
philly_traffic_12_first.head()
```

```python
# Transpose Index from columns, rename columns, and clean up frame to prepare for plotting

chi_traffic_12 = chi_traffic_12_first.T
chi_traffic_12 = chi_traffic_12.rename(columns={4:"Total"})
chi_traffic_12.drop(index=chi_traffic_12.index[0], axis=0, inplace=True)
chi_traffic_12.drop(index=13, axis=0, inplace=True)
chi_traffic_12.drop(labels=[0,1,2,3], axis=1, inplace=True)
chi_traffic_12['Total'] = chi_traffic_12['Total'].astype(float)

chi_traffic_19 = chi_traffic_19_first.T
chi_traffic_19 = chi_traffic_19.rename(columns={4:"Total"})
chi_traffic_19.drop(index=chi_traffic_19.index[0], axis=0, inplace=True)
chi_traffic_19.drop(index=13, axis=0, inplace=True)
chi_traffic_19.drop(labels=[0,1,2,3,5,6,7], axis=1, inplace=True)
chi_traffic_19['Total'] = chi_traffic_19['Total'].astype(float)
```

```python
# Transpose Index from columns, rename columns, and clean up frame to prepare for plotting

ny_traffic_12 = ny_traffic_12_first.T
ny_traffic_12 = ny_traffic_12.rename(columns={4:"Total"})
ny_traffic_12.drop(index=ny_traffic_12.index[0], axis=0, inplace=True)
ny_traffic_12.drop(index=13, axis=0, inplace=True)
ny_traffic_12.drop(labels=[0,1,2,3], axis=1, inplace=True)
ny_traffic_12['Total'] = ny_traffic_12['Total'].astype(float)

ny_traffic_19 = ny_traffic_19_first.T
ny_traffic_19 = ny_traffic_19.rename(columns={4:"Total"})
ny_traffic_19.drop(index=ny_traffic_19.index[0], axis=0, inplace=True)
ny_traffic_19.drop(index=13, axis=0, inplace=True)
ny_traffic_19.drop(labels=[0,1,2,3], axis=1, inplace=True)
ny_traffic_19['Total'] = ny_traffic_19['Total'].astype(float)
```

```python
x1 = np.arange(1,13,1)    
y1 = chi_traffic_12['Total']
plt.plot(x1, y1, label = "2012")
x2 = np.arange(1,13,1)  
y2 = chi_traffic_19['Total']
plt.plot(x2, y2, label = "2019")
plt.xlabel('Month')
plt.ylabel('Amount')
plt.title('Fatal Chicago Traffic Accidents 2012 vs 2019')
plt.legend()
plt.show()
```
