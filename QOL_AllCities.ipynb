```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
csvpath_chi_crime_12_19 = os.path.join('Chicago', 'Chicago Crime Data 2012 and 2019.csv')
csvpath_LA_crime_12_19 = os.path.join('LA Data', 'Cleaned_LA_crime_data_from_2010_to_2019.csv')
csvpath_philly_crime_12 = os.path.join('Philadelphia', 'Philadelphia Crime Data 2012.csv')
csvpath_philly_crime_19 = os.path.join('Philadelphia', 'Philadelphia Crime Data 2019.csv')
```

```python
with open(csvpath_chi_crime_12_19) as csvfile_chi_crime:
    csvreader_chi_crime = csv.reader(csvfile_chi_crime, delimiter=",")
    csv_header = next(csvreader_chi_crime) 
    chi_crime_df_first = pd.DataFrame(csvreader_chi_crime)

with open(csvpath_LA_crime_12_19) as csvfile_LA_crime:
    csvreader_LA_crime = csv.reader(csvfile_LA_crime, delimiter=",")
    csv_header = next(csvreader_LA_crime) 
    LA_crime_df_first = pd.DataFrame(csvreader_LA_crime)
    
with open(csvpath_philly_crime_12) as csvfile_philly_crime_12:
    csvreader_philly_crime_12 = csv.reader(csvfile_philly_crime_12, delimiter=",")
    csv_header = next(csvreader_philly_crime_12) 
    philly_crime12_df_first = pd.DataFrame(csvreader_philly_crime_12)

with open(csvpath_philly_crime_19) as csvfile_philly_crime_19:
    csvreader_philly_crime_19 = csv.reader(csvfile_philly_crime_19, delimiter=",")
    csv_header = next(csvreader_philly_crime_19) 
    philly_crime19_df_first = pd.DataFrame(csvreader_philly_crime_19)
```
