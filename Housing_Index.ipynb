```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
csvpath_housing = os.path.join('All City Data', 'Cleaned_Housing_Price_Index_All_Cities.csv')

# Reading in csv data sets 

with open(csvpath_housing) as csvfile_housing:
    csvreader_housing = csv.reader(csvfile_housing, delimiter=",")
    csv_header = next(csvreader_housing) 
    all_housing_df = pd.DataFrame(csvreader_housing)
```

```python
# Renaming columns and converting year column to integer for filtering purposes

all_housing_df = all_housing_df.rename(columns={0:"City Code", 1:"City", 2:"Year", 3:"Quarter", 4:"Not Seasonally Adjusted Index", 5:"Seasonally Adjusted Index"})
all_housing_df[["Not Seasonally Adjusted Index", "Seasonally Adjusted Index"]] = all_housing_df[["Not Seasonally Adjusted Index", "Seasonally Adjusted Index"]].apply(pd.to_numeric)
all_housing_df['Year'].astype(int)
```

```python
# Creating new dataframes for each city for years 2012 & 2019

chi_housing_2012 = all_housing_df.loc[(all_housing_df["Year"] == "2012") & (all_housing_df["City Code"] == "16984"), :]
LA_housing_2012 = all_housing_df.loc[(all_housing_df["Year"] == "2012") & (all_housing_df["City Code"] == "31084"), :]
philly_housing_2012 = all_housing_df.loc[(all_housing_df["Year"] == "2012") & (all_housing_df["City Code"] == "37964"), :]
ny_housing_2012 = all_housing_df.loc[(all_housing_df["Year"] == "2012") & (all_housing_df["City Code"] == "35614"), :]

chi_housing_2019 = all_housing_df.loc[(all_housing_df["Year"] == "2019") & (all_housing_df["City Code"] == "16984"), :]
LA_housing_2019 = all_housing_df.loc[(all_housing_df["Year"] == "2019") & (all_housing_df["City Code"] == "31084"), :]
philly_housing_2019 = all_housing_df.loc[(all_housing_df["Year"] == "2019") & (all_housing_df["City Code"] == "37964"), :]
ny_housing_2019 = all_housing_df.loc[(all_housing_df["Year"] == "2019") & (all_housing_df["City Code"] == "35614"), :]
```

```python
chi_housing_2012_quarter = chi_housing_2012.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])
LA_housing_2012_quarter = LA_housing_2012.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])
philly_housing_2012_quarter = philly_housing_2012.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])
ny_housing_2012_quarter = ny_housing_2012.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])

chi_housing_2019_quarter = chi_housing_2019.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])
LA_housing_2019_quarter = LA_housing_2019.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])
philly_housing_2019_quarter = philly_housing_2019.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])
ny_housing_2019_quarter = ny_housing_2019.groupby(['Quarter']).mean(["Not Seasonally Adjusted Index","Seasonally Adjusted Index"])
```

```python
# Reading in Weather data to extract quarterly ranges for comparison with this housing data
csvpath_chi_12 = os.path.join('2009,2019 Weather CSV', 'Chicago 2012.csv')
csvpath_chi_19 = os.path.join('2009,2019 Weather CSV', 'Chicago 2019.csv')
csvpath_LA_12 = os.path.join('2009,2019 Weather CSV', 'LAS 2012.csv')
csvpath_LA_19 = os.path.join('2009,2019 Weather CSV', 'LAS 2019.csv')
csvpath_ny_12 = os.path.join('2009,2019 Weather CSV', 'NY 2012.csv')
csvpath_ny_19 = os.path.join('2009,2019 Weather CSV', 'NY 2019.csv')
csvpath_philly_12 = os.path.join('2009,2019 Weather CSV', 'PHL 2012.csv')
csvpath_philly_19 = os.path.join('2009,2019 Weather CSV', 'PHL 2019.csv')
```

```python
with open(csvpath_chi_12) as csvfile_chi_12:
    csvreader_chi_12 = csv.reader(csvfile_chi_12, delimiter=",")
    csv_header = next(csvreader_chi_12) 
    chi_12_df_first = pd.DataFrame(csvreader_chi_12)
    
with open(csvpath_chi_19) as csvfile_chi_19:
    csvreader_chi_19 = csv.reader(csvfile_chi_19, delimiter=",")
    csv_header = next(csvreader_chi_19) 
    chi_19_df_first = pd.DataFrame(csvreader_chi_19)

with open(csvpath_LA_12) as csvfile_LA_12:
    csvreader_LA_12 = csv.reader(csvfile_LA_12, delimiter=",")
    csv_header = next(csvreader_LA_12) 
    LA_12_df_first = pd.DataFrame(csvreader_LA_12)

with open(csvpath_LA_19) as csvfile_LA_19:
    csvreader_LA_19 = csv.reader(csvfile_LA_19, delimiter=",")
    csv_header = next(csvreader_LA_19) 
    LA_19_df_first = pd.DataFrame(csvreader_LA_19)

with open(csvpath_ny_12) as csvfile_ny_12:
    csvreader_ny_12 = csv.reader(csvfile_ny_12, delimiter=",")
    csv_header = next(csvreader_ny_12) 
    ny_12_df_first = pd.DataFrame(csvreader_ny_12)

with open(csvpath_ny_19) as csvfile_ny_19:
    csvreader_ny_19 = csv.reader(csvfile_ny_19, delimiter=",")
    csv_header = next(csvreader_ny_19) 
    ny_19_df_first = pd.DataFrame(csvreader_ny_19)
    
with open(csvpath_philly_12) as csvfile_philly_12:
    csvreader_philly_12 = csv.reader(csvfile_philly_12, delimiter=",")
    csv_header = next(csvreader_philly_12) 
    philly_12_df_first = pd.DataFrame(csvreader_philly_12)

with open(csvpath_philly_19) as csvfile_philly_19:
    csvreader_philly_19 = csv.reader(csvfile_philly_19, delimiter=",")
    csv_header = next(csvreader_philly_19) 
    philly_19_df_first = pd.DataFrame(csvreader_philly_19)
```

```python
# Renaming original 2012 data frame columns and making a new column to store the Months calculated using a datetimeindex function on Date column 

chi_12_df = chi_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
chi_12_df['Quarter'] = pd.DatetimeIndex(chi_12_df['Date']).quarter

LA_12_df = LA_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
LA_12_df['Quarter'] = pd.DatetimeIndex(LA_12_df['Date']).quarter

ny_12_df = ny_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
ny_12_df['Quarter'] = pd.DatetimeIndex(ny_12_df['Date']).quarter

philly_12_df = philly_12_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
philly_12_df['Quarter'] = pd.DatetimeIndex(philly_12_df['Date']).quarter

chi_12_df
```

```python
chi_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = chi_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
chi_12_quarter = chi_12_df.groupby(['Quarter']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

LA_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = LA_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
LA_12_quarter = LA_12_df.groupby(['Quarter']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

ny_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = ny_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
ny_12_quarter = ny_12_df.groupby(['Quarter']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])

philly_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]] = philly_12_df[["Temp Avg", "Precipitation", "Wind Speed", "Max Temp"]].apply(pd.to_numeric)
philly_12_quarter = philly_12_df.groupby(['Quarter']).mean([['Temp Avg', 'Precipitation', 'Wind Speed']])
```

```python
# Renaming original 2019 data frame columns and making a new column to store the Months calculated using a datetimeindex function on Date column 

chi_19_df = chi_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
chi_19_df['Quarter'] = pd.DatetimeIndex(chi_19_df['Date']).quarter

LA_19_df = LA_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
LA_19_df['Quarter'] = pd.DatetimeIndex(LA_19_df['Date']).quarter

ny_19_df = ny_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
ny_19_df['Quarter'] = pd.DatetimeIndex(ny_19_df['Date']).quarter

philly_19_df = philly_19_df_first.rename(columns={0:"Date", 1:"Temp Avg", 2:"Temp Min", 3:"Max Temp",4:"Precipitation", 5:"Snow", 6:"Wind Speed"})
philly_19_df['Quarter'] = pd.DatetimeIndex(philly_19_df['Date']).quarter

```

```python
# Create various plots of data for analysis

x = np.arange(1,5,1)
x2 = np.arange(1,5,1)
fig, axs = plt.subplots(2, 2)  

y = chi_housing_2012_quarter['Not Seasonally Adjusted Index']
axs[0, 0].plot(x,y, label = "2012") 
y2 = chi_housing_2019_quarter['Not Seasonally Adjusted Index']
axs[0, 0].plot(x2,y2, label = "2019")
axs[0, 0].set_title("Chicago Housing Index Averages 2012 vs 2019")
axs[0, 0].legend(loc='upper right')
axs[0, 0].set_ylabel('Price')

y = LA_housing_2012_quarter['Not Seasonally Adjusted Index']
axs[1, 0].plot(x,y, label = "2012")
y2 = LA_housing_2019_quarter['Not Seasonally Adjusted Index']
axs[1, 0].plot(x2,y2, label = "2019")
axs[1, 0].set_title("Los Angeles Housing Index Averages 2012 vs 2019")
axs[1, 0].legend(loc='upper right')
axs[1, 0].set_ylabel('Price')

y = ny_housing_2012_quarter['Not Seasonally Adjusted Index']
axs[0, 1].plot(x,y, label = "2012")
y2 = ny_housing_2019_quarter['Not Seasonally Adjusted Index']
axs[0, 1].plot(x2,y2, label = "2019")
axs[0, 1].set_title("New York Housing Index Averages 2012 vs 2019")
axs[0, 1].legend(loc='upper right')
axs[0, 1].set_ylabel('Price')

y = philly_housing_2012_quarter['Not Seasonally Adjusted Index']
axs[1, 1].plot(x,y, label = "2012")
y2 = philly_housing_2019_quarter['Not Seasonally Adjusted Index']
axs[1, 1].plot(x2,y2, label = "2019")
axs[1, 1].set_title("Philadelphia Housing Index Averages 2012 vs 2019")
axs[1, 1].legend(loc='upper right')
axs[1, 1].set_ylabel('Price')

plt.subplots_adjust(left=0,
                    bottom=3, 
                    right=2, 
                    top=5, 
                    wspace=0.4, 
                    hspace=0.3)
plt.setp(axs, xticks=np.arange(1,5,1), xticklabels=['Q1', 'Q2', 'Q3', 'Q4'])

for ax in axs:
    plt.ylim(0, max(y2)+15)
```
